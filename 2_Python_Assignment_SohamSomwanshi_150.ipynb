```python
today = date.today() #use the datetime for current date
```

```python
search_term = ['aap','bsp','bjp','sp','congress'] #list[aap,bsp,bjp,sp,congress]
from_date = '2021-01-01' 
max=250 #max search for each item from list
```

1.Create a function to extract tweets from a given
list[aap,bsp,bjp,sp,congress] each corresponding  txt  file need to have name  or ex- aap.txt

```python
def create(search_term):
     for i,search_term in enumerate(search_term):#use enumerate function for passing the every list item
       st=(search_term +".txt")
       os.system("snscrape --format '{content!r}'"+f" --max-results {max} --since {from_date} twitter-search '{search_term} until:{today}' > {st}")#Scfrapdata from tweeter using snScrap module
       if os.stat(st).st_size == 0: #use condition for checking count
         counter = 0
       else:
         df = pd.read_csv(st, names=['content'])
         counter = df.size
       print('Number Of Tweets : '+ str(counter))
ss=create(search_term)
```

2.Above create function will create n number of txt files merge them into one txt file and convert that file 
into CSV data needs to be under header ‘text’

```python
"""2.Above create function will create n number of txt files merge them into one txt file and convert that file 
into CSV data needs to be under header ‘text’"""   
filenames = ["aap.txt", "bjp.txt", "bsp.txt","congress.txt","sp.txt"]
with open("mergefile.txt", "w") as mf: #Create a Function For merge a data 
    for i in filenames:
        with open(i) as fn:
            contents = fn.read()
            mf.write(contents)
df = pd.read_csv('mergefile.txt', names=['content'])
df.to_csv('main.csv' , header='text' , index =True) #convert merge file into .csv file
```

3.Cleaning of data :-
•	Remove punctuation 
•	Remove username 
•	Remove emoji

```python
"""3.Cleaning of data :-
•	Remove punctuation 
•	Remove username 
•	Remove emoji"""
!pip install emoji 
import re
import emoji 
with open("final.txt", "w") as fl: #create new clean data add
  with open("mergefile.txt", "r+") as mf:
    for row in mf:
      row = mf.read()
      row=str(row)
      #Remove User Name 
      Tweet = re.sub('@[\w]+','',row)
      #Remove emoji
      Tweet=emoji.get_emoji_regexp().sub(r'', Tweet)
      punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
      for ele in Tweet: #using for loop pass each element of tweet
        if ele in punc: 
          Tweet = Tweet.replace(ele, "") #After checking Remove or we say replace puctuation with ""
      fl.write(Tweet) #write clear tweet in final file
      print(Tweet)   
  mf.close()
fl.close()
df = pd.read_csv('final.txt', names=['content'])
df.to_csv('main.csv', header='text', index =False) #import final clean data file into  .csv file

```
